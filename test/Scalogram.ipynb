```julia
f=1./period;
ax=subplot(111);
cf=ax[:contourf](t,f,log(float(abs(wave).^2)),50);
cb=colorbar(cf);
cb[:set_label]("Wavelet Power");
ax[:fill_between](t,1./coi, color="white", alpha=0.5);
yscale("log");
xlabel("Time [s]");
ylabel("Frequency [Hz]");
ylim([minimum(f),maximum(f)]);
```
